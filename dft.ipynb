```python
import numpy as np
import matplotlib.pyplot as plt

#dft X[k]=sum(x[n]exp(-2pi*n*k/N))|n=N-1 n=0
N=100
frequency = 100
sampling_rate = 4000
n = np.arange(int(N)+1)
x = np.cos(2*np.pi*frequency*n/sampling_rate)
plt.stem(n,x,linefmt='', markerfmt='o', basefmt='')
```

```python
#dft X[k]=sum(x[n]exp(-j2pi*n*k/N))|n=N-1 n=0
M=100
k = np.arange(int(M)+1)
X = np.array([ np.sum( x * np.exp(-2*np.pi*1j*n/N*m) ) for m in k])
mag_X = np.abs(X)

plt.stem(k,mag_X)
```

```python
phase_X =np.angle(X)
plt.stem(k,phase_X)
```
