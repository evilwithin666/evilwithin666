```python
import numpy as np 
import matplotlib.pyplot as plt

amplitude = 1.0
duration = 4.0
time_period = 1.0
sampling_rate = 1000

time = np.linspace(-duration/2,duration/2,int(duration*sampling_rate)+1,endpoint=True)

def pulse(x,start,end):
    return 1 if start<=x<=end else 0

rect_pulse = np.array([pulse(t,-time_period/2,time_period/2) for t in time])
plt.plot(time, rect_pulse)
plt.show()
```

```python
# fourier transform = integral( x(t) exp(-jwt) dt )

dt = 1/sampling_rate

#define frequencies
max_freq = 5
freq_samples = 1000
frequencies = np.linspace(-max_freq, max_freq, freq_samples+1, endpoint=True)

#complex_exp = np.exp(-1j*2*np.pi*f*time)

def ft(f):
     return np.sum( rect_pulse * np.exp(-1j*2*np.pi*f*time) * dt)

f_trans = np.array([ft(f) for f in frequencies])
mag_ft = np.array([np.abs(ft(f)) for f in frequencies])
phase_ft = np.array([np.angle(ft(f)) for f in frequencies])

print(f_trans)
plt.plot(frequencies,np.real(f_trans))

```
